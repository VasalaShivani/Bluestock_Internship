```python
import pandas as pd
import os

# Automatically detect the correct folder path
data_dir = '../data/raw/' if os.path.exists('../data/raw/') else 'data/raw/'
nav_file = f'{data_dir}02_nav_history.csv'

try:
    print("--- Starting NAV History Processing ---")
    
    # 1. Load ONLY the NAV History file
    nav_history = pd.read_csv(nav_file)
    print(f"✅ SUCCESS: '{nav_file}' loaded perfectly!\n")
    
    # 2. Display a summary of the data
    print("--- NAV History Summary ---")
    print(f"Total Records: {len(nav_history)}")
    print(f"Columns available: {list(nav_history.columns)}\n")
    
    # 3. Check for any missing data
    missing_values = nav_history.isnull().sum()
    
    if missing_values.sum() == 0:
        print("✅ Data Quality Check Passed: No missing values found.")
        print("✅ Your NAV History data is clean and ready for analysis!")
    else:
        print("⚠️ Warning: Missing values detected in the following columns:")
        print(missing_values[missing_values > 0])
        
except FileNotFoundError:
    print(f"❌ Error: Could not find '{nav_file}'. Please check that the file is in the data/raw/ folder.")
except Exception as e:
    print(f"❌ Unexpected Error: {e}")
```
