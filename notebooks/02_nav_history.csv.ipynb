```python
import pandas as pd
import os

# Automatically detect the correct folder path
data_dir = '../data/raw/' if os.path.exists('../data/raw/') else 'data/raw/'

# --- TASK 7: VALIDATE AMFI CODES ---
master_file = f'{data_dir}01_fund_master.csv'
nav_file = f'{data_dir}02_nav_history.csv'

try:
    # 1. Load BOTH files right here so we don't rely on Task 6's memory
    fund_master = pd.read_csv(master_file)
    nav_history = pd.read_csv(nav_file)
    print(f"✅ Task 7: {nav_file} loaded successfully!\n")
    
    # 2. Run the validation
    if 'amfi_code' in fund_master.columns and 'amfi_code' in nav_history.columns:
        master_codes = set(fund_master['amfi_code'])
        nav_codes = set(nav_history['amfi_code'])
        
        missing_in_nav = master_codes - nav_codes
        
        print("--- Data Quality Report ---")
        print(f"Total unique schemes in Fund Master: {len(master_codes)}")
        print(f"Total unique schemes in NAV History: {len(nav_codes)}\n")
        
        if len(missing_in_nav) == 0:
            print("✅ Validation Passed: All AMFI codes in fund_master exist in nav_history.")
        else:
            print(f"⚠️ Warning: Found {len(missing_in_nav)} codes in fund_master missing from nav_history.")
    else:
        print("❌ Error: The column 'amfi_code' is missing from one of your files.")
        
except FileNotFoundError as e:
    print(f"❌ File Error: {e}")
except Exception as e:
    print(f"❌ Unexpected Error in Task 7: {e}")
    
```
