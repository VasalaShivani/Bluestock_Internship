```python
import pandas as pd
import os

# Automatically detect the correct folder path
data_dir = '../data/raw/' if os.path.exists('../data/raw/') else 'data/raw/'

# --- TASK 6: FUND MASTER ---
# FIXED: Added the '01_' to perfectly match your computer's file!
master_file = f'{data_dir}01_fund_master.csv' 

try:
    # Load the CSV
    fund_master = pd.read_csv(master_file)
    print(f"✅ Task 6: {master_file} loaded successfully!\n")
    
    print("--- Fund Master EDA ---")
    
    if 'fund_house' in fund_master.columns:
        print("Unique Fund Houses:\n", fund_master['fund_house'].unique()[:5], "... [Truncated]\n")
        
    if 'category' in fund_master.columns:
        print("Unique Categories:\n", fund_master['category'].unique(), "\n")
        
    if 'sub_category' in fund_master.columns:
        print("Unique Sub-Categories:\n", fund_master['sub_category'].unique()[:5], "... [Truncated]\n")
        
    # Safety check for the risk_grade column
    if 'risk_grade' in fund_master.columns:
        print("Unique Risk Grades:\n", fund_master['risk_grade'].unique(), "\n")
    else:
        print("Unique Risk Grades:\n [Skipped: 'risk_grade' column does not exist in this CSV]\n")
        
except FileNotFoundError:
    print(f"❌ Error: Could not find '{master_file}'. Please check that the file is in the data/raw/ folder.")
except Exception as e:
    print(f"❌ Unexpected Error in Task 6: {e}")
```
